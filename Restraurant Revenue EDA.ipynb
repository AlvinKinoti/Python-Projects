```python
## Visualizing the dataset

# Correlation between variables

revenue = df.drop(["Cuisine_Type"], axis = 1)

correlation = revenue.corr()

plt.figure(figsize=(7,7))
sns.heatmap(correlation)
plt.show()
```

```python
## Checking the most demanded Cuisine type

plt.figure(figsize = (5,5))
plt.barh(df["Cuisine_Type"], df["Number_of_Customers"])
plt.xlabel("Demand")
plt.ylabel("Cuisine Type")
plt.grid(True)
plt.title("Demand per Cuisine Type")
plt.show()
```

```python
## Histogram for the dataset

plt.figure(figsize=(7,7))
plt.hist(revenue)
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.legend(["Number_of_Customers", "Menu_Price", "Marketing_Spend", "Average_Customer_Spending",	
            "Promotions", "Reviews", "Monthly_Revenue"])
plt.grid()
plt.show()
```
