# Importing the libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Loading the dataset

```python
df = pd.read_csv("bajaj-2003-2020.csv")
df
```

## Exporatory Data Analysis

```python
df.drop(["Symbol"], axis = 1)
```

## b) Cleaning the dataset

```python
df.shape
```

```python
df.describe()
```

### Checking for any empty values

```python
df.isnull().sum()
```

### Selecting only the columns needed 

```python
df1 = df[["Prev Close", "Open Price", "High Price", "Low Price", "Last Price", "Close Price", "Average Price"]]
df1
```

### Checking for any outliers

```python
sns.boxplot([df1["Average Price"]])
plt.grid()
```

### Removing the outliers detected in the Average Price Column.

```python
filter = df["Average Price"]> 500
df_outlier_rem = df[filter]
df_outlier_rem.drop(["Symbol"], axis = 1)
```

# Visualizing the Data

```python
df2 = df_outlier_rem.drop(["Symbol"], axis = 1)
df3 = df2[["Prev Close", "Open Price", "High Price", "Low Price", "Last Price", "Close Price", "Average Price"]]
```

## a) Histogram

```python
plt.figure(figsize = (10,10))
plt.hist(df3)
plt.legend(["Prev Close", "Open Price", "High Price", "Low Price", "Last Price", "Close Price", "Average Price"])
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.grid()
plt.show()
```

# Descriptive Statistics 

```python
df4 = df3[["Prev Close", "Open Price", "High Price", "Low Price", "Last Price", "Close Price", "Average Price"]]
df4
```

## Checking for correlation between selected variables.

```python
df4.corr()
```

# Coming up with a multiple linear regression model

```python
df5 = df4.drop(["Average Price"], axis = 1)
```

### Selecting the X and y values 

```python
X = df5.drop(["Close Price"], axis =1)
y = df5["Close Price"]
```

### Selecting and fitting the linear model

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X,y)
```

### Predicting the Linear Regression Model

```python
lr.predict(X)
```

```python
pred = lr.predict(X)
```

### Plotting the predicted values against the actual values

```python
plt.figure(figsize = (10,10))
plt.scatter(pred, y, s = 250, color = "green")
plt.xlabel("Predicted Value")
plt.ylabel("Close Price")
plt.grid()
plt.title("Predicted Value vs Close Price")
plt.show()
```

### Regression values

```python
from sklearn.metrics import r2_score
```

```python
r2_score(pred,y)
```

#### Conclusion:

The model is a good fit
