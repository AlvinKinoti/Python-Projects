### Cleaning the dataset

```python
df.duplicated().sum()
```

```python
df1 = df.drop(["Unnamed: 0", "Order ID"],axis = 1)
df1
```

#### Dealing with outliers

```python
sns.boxplot(df1["Sales"])
plt.grid(True)
```

```python
# dealing with outliers in the Sales column

q1 = df1["Sales"].quantile(0.25)
q3 = df1["Sales"].quantile(0.75)
```

```python
iqr = q3 - q1
```

```python
print(q1,q3,iqr)
```

```python
upper_limit = q3 + (1.5*iqr)
lower_limit = q1 - (1.5*iqr)

lower_limit, upper_limit
```

```python
df1.loc[(df1["Sales"] > upper_limit) | (df1["Sales"] < lower_limit)]
```

```python
df2 = df1.loc[(df1["Sales"] < upper_limit) & (df1["Sales"] > lower_limit)]
```

```python
df2.shape
```

```python
print("before removing outliers:", len(df1))
print("after removing outliers:", len(df2))
print("outliers:", len(df1)-len(df2))
```

```python
# dealing with outliers in the Price Each column

q1 = df2["Price Each"].quantile(0.25)
q3 = df2["Price Each"].quantile(0.75)

iqr = q3 - q1
```

```python
lower_limit = q1 - (1.5 * iqr)
upper_limit = q3 + (1.5 * iqr)
```

```python
df2.loc[(df2["Price Each"] > upper_limit) | (df2["Price Each"] < lower_limit)]
```

```python
df3 = df2.loc[(df2["Price Each"] < upper_limit) & (df2["Price Each"] > lower_limit)]
df3
```

```python
df3.shape
```

```python
print("before removing outliers:", len(df2))
print("after removing outliers:", len(df3))
print("outliers:", len(df2)-len(df3))
```

### Exploratory Data Analysis

```python
sales = df3
```

```python
sales.corr()
```

#### Histogram for sales 

```python
plt.figure(figsize=(7,7))
plt.hist(sales["Sales"], color = "green", bins = 10)
plt.xlabel("Profit")
plt.ylabel("Frequency")
plt.grid(True)
plt.title("Profit Frequency")
plt.show()
```

#### Histogram for Cost

```python
plt.figure(figsize=(7,7))
plt.hist(sales["Price Each"], color = "red", bins = 10)
plt.xlabel("Cost")
plt.ylabel("Frequency")
plt.grid(True)
plt.title("Cost Frequency")
plt.show()
```

#### Profit Distribution per City

```python
plt.figure(figsize=(7,7))
plt.barh(sales["City"], sales["Sales"])
plt.ylabel("City")
plt.xlabel("Profit")
plt.grid(True)
plt.title("Profit Distribution per City")
plt.show()
```

#### Number of Orders Per Product

```python
plt.figure(figsize=(7,7))
plt.barh(sales["Product"], sales["Quantity Ordered"])
plt.ylabel("Products")
plt.xlabel("Number of Orders")
plt.grid(True)
plt.title("Number of Orders Per Product")
plt.show()
```
