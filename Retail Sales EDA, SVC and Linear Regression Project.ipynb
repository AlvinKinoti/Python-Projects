## Coming up with an SVC Model

```python
import sklearn
```

```python
from sklearn import svm
```

```python
from sklearn.svm import SVC
```

```python
X = df3.drop(["Total Amount"], axis = 1)
y = df3["Total Amount"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train,y_test = train_test_split(X, y, test_size=0.2, random_state =0)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc= StandardScaler()
```

```python
X_train = sc.fit_transform(X_train)
X_test =sc.transform(X_test)
```

```python
clf = svm.SVC()
```

```python
clf.fit(X_train, y_train)
```

### Predicting the model using the classifier

```python
pred_clf = clf.predict(X_test)
```

```python
pred_clf
```

### Checking the accuracy of the model

```python
sklearn.metrics.accuracy_score(y_test, pred_clf)
```

#### The svm classifier is doing a fairly good job at predicting the model

### Generating a classification Building Report

```python
sklearn.metrics.classification_report(y_test, pred_clf)
```

```python
print(sklearn.metrics.classification_report(y_test, pred_clf))
```
